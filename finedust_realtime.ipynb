```
rows =[]

for item in ItemList:
    a = item.find('datatime').text
    g = item.find('pm10value').text
    rows.append({"datatime": a, 
                 "pm10value": g})
columns = ["datatime", "pm10value"]
df = pd.DataFrame(rows, columns=columns)
print(df)

df.to_csv("pm10.csv")
```
